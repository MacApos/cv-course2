```python
img1 = cv2.imread(filename='view.jpg')
img2 = cv2.imread(filename='bear.jpg')
```

```python
cv2_imshow(img2)
```

```python
# Łączenie obrazów
add_img = cv2.add(src1=img1, src2=img2)
add_img.shape
```

```python
cv2_imshow(add_img)
```

```python
# Blendowanie zdjęć
dst = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
cv2_imshow(dst)
```
